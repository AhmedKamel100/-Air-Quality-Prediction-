#Importing Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Cleaning the Dataset

```python
dataset = pd.read_excel('AirQualityUCI.xlsx')
X = dataset.iloc[:, 3:15]
y = dataset.iloc[:, 2]

for col in X.columns:
    mean_val = X[X[col] != -200][col].mean()
    X[col] = X[col].replace(-200, mean_val)

y_mean = y[y != -200].mean()
y = y.replace(-200, y_mean)

X = X.values
y = y.values
```

# Trian the Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Test The Model

```python
y_pred = regressor.predict(X_test)

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# Graph


```python
plt.figure(figsize=(10, 6))
plt.plot(y_test[:100], label='Actual', marker='o')
plt.plot(y_pred[:100], label='Predicted', marker='x')
plt.title('Actual vs Predicted CO values')
plt.xlabel('Sample Index')
plt.ylabel('CO Concentration')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
